# Projeto 1 — Análise de Rentabilidade

## Objetivo
Avaliar margem de lucro e identificar os produtos mais rentáveis.

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
df = pd.read_csv("dataset_rentabilidade.csv")
df.head()
```

```python
df["receita"] = df["quantidade_vendida"] * df["preco_unitario"]
df["custo_total"] = df["quantidade_vendida"] * df["custo_unitario"]
df["lucro"] = df["receita"] - df["custo_total"]

df.head()
```

```python
lucro_produto = df.groupby("produto")["lucro"].sum()

lucro_produto.plot(kind="bar")
plt.title("Lucro por Produto")
plt.show()
```

```python
df["margem_%"] = (df["lucro"] / df["receita"]) * 100

margem_media = df["margem_%"].mean()
margem_media
```

# Principais Insights

- A **Calça Jeans** apresentou o maior lucro absoluto no período analisado, indicando forte desempenho comercial aliado a boa margem unitária;
- A margem média da empresa foi de aproximadamente **45,7%**, demonstrando uma operação com boa eficiência de custos;
- A categoria **Calçados** apresentou maior rentabilidade proporcional, sugerindo potencial para expansão de mix ou aumento de estoque estratégico;
- Produtos com menor margem podem ser avaliados para renegociação com fornecedores ou ajuste de preço.
