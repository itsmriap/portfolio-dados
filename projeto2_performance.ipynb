```python
# Projeto 2 — Performance Comercial

## Objetivo
Analisar desempenho de vendedores comparando meta e resultado.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset_performance.csv")
df.head()
```

```python
df["atingimento_%"] = (df["valor_venda"] / df["meta_mensal"]) * 100
df.head()
```

```python
ranking = df.groupby("vendedor")["valor_venda"].sum().sort_values(ascending=False)

ranking.plot(kind="bar")
plt.title("Ranking de Vendedores")
plt.show()
```

```python
regiao = df.groupby("regiao")[["meta_mensal", "valor_venda"]].sum()
regiao.plot(kind="bar")
plt.title("Meta vs Realizado por Região")
plt.show()
```

# Principais Insights

- O vendedor com maior faturamento no período foi **Carlos Mendes**, demonstrando alta capacidade de conversão e superação de metas;
- A região **Sudeste** apresentou o maior volume total de vendas, indicando maior concentração de mercado e potencial de expansão comercial;
- A média geral de atingimento de meta foi superior a 100% em algumas regiões, sugerindo metas conservadoras ou forte desempenho da equipe;
- Vendedores com percentual abaixo da meta podem demandar ações de treinamento ou revisão de estratégia comercial;
- Existe oportunidade de replicar as práticas dos vendedores de alta performance para elevar o resultado médio da equipe.
